```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.utils import plot_model
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error
```

```python
np.random.seed(42)
X = np.random.rand(100, 1)
y = 2 * X + 1 + 0.1 * np.random.randn(100, 1)
```

```python
# Creazione del plot
plt.figure(figsize=(8, 6))
plt.scatter(X, y, color='blue', label='Dati')
plt.plot(X, 2 * X + 1, color='red', label='Linea Reale: y = 2X + 1')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Dati generati per la regressione lineare')
plt.legend()
plt.grid(True)
plt.show()
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Build a linear regression model
# Definizione del modello per la regressione lineare
model = keras.Sequential([
    layers.InputLayer(input_shape=(1,)),  # Definisce esplicitamente l'input shape
    layers.Dense(1, name='output_layer')  # Output layer
])

# Compile the model
model.compile(
    optimizer='adam', # Optimizer: algoritmo di ottimizzazione
    loss='mse' # mean squared error: errore quadratico medio
)

# Train the model
model.fit(X_scaled, y, epochs=1200, verbose=1)
```

```python
# Make predictions
predictions = model.predict(X_scaled)

# Plot the results
plt.figure(figsize=(8, 6))

plt.scatter(X, y, color='blue', label='Dati')
plt.plot(X, 2 * X + 1, color='red', label='Linea Reale: y = 2X + 1')
plt.plot(X, predictions, color='darkred', label='Linea Predetta')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Regressione Lineare con TensorFlow')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Generazione dei dati
np.random.seed(42)
X = np.linspace(0, 2 * np.pi, 250)
y = np.sin(X) + 0.3 * np.random.randn(250)

# Standardizzazione dei dati
scaler = StandardScaler()
X = scaler.fit_transform(X.reshape(-1, 1)).reshape(-1)
y = scaler.fit_transform(y.reshape(-1, 1)).reshape(-1)

# Divisione dei dati in train e validation set in modo casuale
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.3, random_state=42)

# Definizione del modello
model = keras.Sequential([
    layers.Dense(128, input_dim=1, activation='sigmoid'),
    layers.Dense(128, activation='sigmoid'),
    layers.Dense(128, activation='sigmoid'),
    layers.Dense(1)
])

# Compilazione del modello
model.compile(optimizer='adam', loss='mean_squared_error')

# Addestramento del modello
history = model.fit(X_train, y_train, epochs=500, validation_data=(X_val, y_val), verbose=1)

# Previsioni
y_train_pred = model.predict(X_train)
y_val_pred = model.predict(X_val)
```

```python
model.summary()
```

```python
# Sort X and y values by X values for better plotting
sort_train_order = X_train.argsort()
X_train = X_train[sort_train_order]
y_train = y_train[sort_train_order]

sort_val_order = X_val.argsort()
X_val = X_val[sort_val_order]
y_val = y_val[sort_val_order]


y_train_pred = y_train_pred[sort_train_order]
y_val_pred = y_val_pred[sort_val_order]

# Plot dei dati e delle previsioni
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.scatter(X_train, y_train, label='Train Data', color='blue')
plt.scatter(X_val, y_val, label='Validation Data', color='green')
plt.plot(X_train, y_train_pred, label='Train Predictions', color='red')
plt.plot(X_val, y_val_pred, label='Validation Predictions', color='orange')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Dati e Previsioni')
plt.legend()

# Plot della perdita durante l'addestramento
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.xlabel('Epoche')
plt.ylabel('Loss')
plt.title('Andamento della Loss durante l\'Addestramento')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# Definizione del modello con maggiore complessità
model = keras.Sequential([
    layers.Dense(1024, input_dim=1, activation='relu'),
    layers.Dense(512, activation='relu'),
    layers.Dense(512, activation='sigmoid'),
    layers.Dense(512, activation='relu'),
    layers.Dense(1)
])

# Compilazione del modello
model.compile(optimizer='adam', loss='mean_squared_error')

# Addestramento del modello per molte epoche
history = model.fit(X_train, y_train, epochs=200, validation_data=(X_val, y_val), verbose=1)
```
