```python
import numpy as np
import matplotlib.pyplot as plt

# Input range per le funzioni di attivazione
x = np.linspace(-10, 10, 500)

# Definizione delle funzioni di attivazione
def linear(x):
    return x

def relu(x):
    return np.maximum(0, x)

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)

# Calcolo dei valori per ciascuna funzione
y_linear = linear(x)
y_relu = relu(x)
y_sigmoid = sigmoid(x)
y_tanh = tanh(x)

# Grafico delle funzioni
plt.figure(figsize=(12, 8))

# Funzione Lineare
plt.subplot(2, 2, 1)
plt.plot(x, y_linear, label='Linear', color='blue')
plt.title("Funzione Lineare")
plt.xlabel("x")
plt.ylabel("y")
plt.axhline(0, color='black', linestyle='--', linewidth=0.7)
plt.axvline(0, color='black', linestyle='--', linewidth=0.7)
plt.grid(True)
plt.legend()

# Funzione ReLU
plt.subplot(2, 2, 2)
plt.plot(x, y_relu, label='ReLU', color='orange')
plt.title("Funzione ReLU")
plt.xlabel("x")
plt.ylabel("y")
plt.axhline(0, color='black', linestyle='--', linewidth=0.7)
plt.axvline(0, color='black', linestyle='--', linewidth=0.7)
plt.grid(True)
plt.legend()

# Funzione Sigmoide
plt.subplot(2, 2, 3)
plt.plot(x, y_sigmoid, label='Sigmoid', color='green')
plt.title("Funzione Sigmoide")
plt.xlabel("x")
plt.ylabel("y")
plt.axhline(0, color='black', linestyle='--', linewidth=0.7)
plt.axvline(0, color='black', linestyle='--', linewidth=0.7)
plt.grid(True)
plt.legend()

# Funzione Tanh
plt.subplot(2, 2, 4)
plt.plot(x, y_tanh, label='Tanh', color='red')
plt.title("Funzione Tanh")
plt.xlabel("x")
plt.ylabel("y")
plt.axhline(0, color='black', linestyle='--', linewidth=0.7)
plt.axvline(0, color='black', linestyle='--', linewidth=0.7)
plt.grid(True)
plt.legend()

# Mostra tutti i grafici
plt.tight_layout()
plt.show()
```
